# 2. Weekly graph question

Below are the graphs from last week's weekly graph question.  Suppose you wanted to cause the viewer to examine the last three months of data first.  How could you use preattentive attributes to do that?  Suggest several possibilities.

```python
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
numdates = 12
np.random.seed(0)
time_series = 100 * (np.random.normal(size = numdates) / 5 + np.arange(numdates) / 16)
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Nitrate level (% of eutrophic value)")
```

```python
np.random.seed(1)
time_series = 100 * (1 - (np.random.normal(size = numdates) / 8 + np.arange(numdates) / 16))
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Phosphate level (% of eutrophic value)")
```

To draw attention to the last three months of data, I would use several preattentive attributes. I would highlight the last three months using a contrasting color while coloring the rest of the line in with a bland color. This makes sure the last three months stand out. Increasing the line thickness or marker size for tyhe three months to emphasize importance is another idea. Adding shape changes could be useful for gertting those points to stand out even more. Interaction or animation like motion or blinking effects could also help that would further enhance visibility. These design elements can help in guiding the viewers eyes without requiring effort.

# 4. Storytelling With Data plot

Reproduce any graph of your choice in p. 99-119 of the Storytelling With Data book as best you can. (The first half of chapter four). You do not have to get the exact data values right, just the overall look and feel.

```python
import matplotlib.pyplot as plt
import pandas as pd

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", 
          "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
received = [160, 185, 250, 150, 190, 170, 150, 200, 165, 140, 155, 180]
processed = [160, 185, 248, 150, 190, 165, 140, 160, 140, 110, 130, 160]

df = pd.DataFrame({
    "Month": months,
    "Received": received,
    "Processed": processed
})

plt.figure(figsize=(10, 5))
plt.plot(df["Month"], df["Received"], label="Received", color="steelblue", linewidth=3)
plt.plot(df["Month"], df["Processed"], label="Processed", color="firebrick", linewidth=3)

plt.xticks(rotation=0)
plt.ylim(0, 300)
plt.ylabel("")
plt.xlabel("")
plt.legend(loc="center left", bbox_to_anchor=(1, 0.5))
plt.title("")
plt.tight_layout()

plt.show()
```

This graph is from page 113, figure 4.10 of the Storytelling with Data textbook. This graph compares the number of tickets received to the number of tickets processed each month. The use of the two distinct colors makes it easy to differentiate between the two trend lines. For the part where the lines overlap, it highlight when the number of tickets processed was lagging behind the number received. Simple and clean, the graph emphasizes comparison over time, which makes it a lot easier for viewers to find patterns, spikes, or backlogs in ticket handling.
